**What specific factors affect student grades**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
s_df = pd.read_csv("student_data.csv")
print(s_df.columns)
pd.set_option('display.max_columns', None)
s_df.head()
```

No data is missing

```python
s_df.isnull().sum()
```

Column for average grade

```python
s_df["avg_grade"] = (s_df.G1+s_df.G2+s_df.G3)/3 
s_df["avg_grade"]
```

***Grades do not seem to be effected by gender***

```python
sns.barplot(x="sex", y="avg_grade", data=s_df, palette="rocket")
plt.plot()
```

**How parents' education impact grades**


The father eduaction seems to be kind of mixed in affecting the grade, while you can see the top mother education of 4 is actually led to an increase in high grades. Meaning higher mother education affects the student's grades positively 

```python
sns.catplot(x="Fedu", y="avg_grade", kind="swarm", data=s_df)
plt.show()
sns.catplot(x="Medu", y="avg_grade", kind="swarm", data=s_df)
plt.show()
```

**How family relationships impact grades**

Lower family relationships with females make them tend to have lower grades than males. After a relationship rating of 3 (the middle), its pretty much the same. This can mean that grades of females suffer more in a low family relationships

```python
sns.boxplot(x="famrel", y="avg_grade", data=s_df, hue="sex", palette="rocket")

plt.legend(loc=0) #reposition legend
plt.show()
```

**How family status impacts grades**


Family status doesn't seem to have much of an impact, although when Pstatus = A, some grades of females are lower than males. 

```python
sns.violinplot(x="Pstatus", y="avg_grade", hue="sex", data=s_df, palette="rocket")
plt.show()
```

**How time spent away from school impacts**

The amount of times students go out does not seem to affect grades much. Students with no freetime at all have slightly lower grades. It indicates that travel more time does affects the grade negatively, although its debatable since there is less data on people who have lots of travel time

```python
sns.violinplot(x="goout", y="avg_grade", data=s_df)
plt.show()
sns.boxplot(x="freetime", y="avg_grade", data=s_df)
plt.show()
sns.catplot(x="traveltime", y="avg_grade", data=s_df)
plt.show()
```

## Conclusion

**Specific factors affect student grades**

- Gender has no impact on student performance
- High education from a student's mother make them more likely to perform better
- Low family relationship's will likely decrease a female's grades
- Family status generally has no affect on a grade, but for some females, a family status of "A" will leave them with longer grades than some males
- Time spent away from school: Students going out has no affect on the grade, Students will very less freetime appear to have slightly lower grades than others, and the more travel time a student has, the grades seem to go lower. (Although there is not enough data to fully support this)
